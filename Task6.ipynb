```python
#importing Required Libraries
import sklearn.datasets as datasets
import pandas as pd
```

```python
#importing available datasets
iris=datasets.load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
#displaying top records
X.head()
```

```python
#displaying end records
X.tail()
```

```python
#description of the available datasets about their data type,not null columns
X.info()
```

```python
#describing datasets on the basis of count,mean,std,min and max
X.describe()
```

```python
#checking if data contains null constraint too
X.isnull().sum()
```

```python
#targeting the data
Y = iris.target
Y
```

```python
#importing required library to split and train data accordingly
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
#Defining the Decision Tree Algorihthm
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)

print('Decision Tree Classifer Created Successfully')
```

```python
y_predict = dtc.predict(X_test)
```

```python
#Constructing confusion Matrix according to Decision Tree Data
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# finally,Visualizing the Decision Tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(dtc, feature_names = fn, class_names = cn, filled = True);
```
